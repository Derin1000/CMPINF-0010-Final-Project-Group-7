### Bus Stops per Neighborhood
There are 6,218 differeent bus stops in the Pittsburgh region (according to WPRDC's Monthly_Updating_Bus_Stop_Usage dataset). Additionally, the dataset provided didn't provide the neighborhoods these stops were located in, just their geographic coordinates. To avoid manually assigning thousands of bus stops to neighborhoods manually, I overlayed the bus stop dataset on top of a map of Pittsburgh neighborhoods using geopandas and used geopandas functionality to count the number of points (bus stops) that fell within each neighborhood.

```python
import geopandas as gpd
from shapely.geometry import Point


bus_data = pd.read_csv("Final Project - Bus Data/Monthly_Updating_Bus_Stop_Usage.csv")

neighborhoods = gpd.read_file("Final Project - Bus Data/Neighborhoods_.shp")

geometry = [Point(xy) for xy in zip(bus_data['longitude'], bus_data['latitude'])]
gdf = gpd.GeoDataFrame(bus_data, geometry=geometry)
gdf.set_crs("EPSG:4326", inplace=True)
gdf = gdf.drop_duplicates(subset="geometry")

fig, ax = plt.subplots(figsize=(20,16))

neighborhoods.plot(ax=ax,edgecolor='black')
gdf.plot(ax=ax, color="red", markersize=0.1)

stops_per_neighborhood = gpd.sjoin(gdf, neighborhoods, how="inner", predicate="within")

counts = stops_per_neighborhood.groupby("hood").size().reset_index(name="point_count")

print("Top 30 Neighborhoods with most bus stops")
print(counts.sort_values(by="point_count", ascending=False).head(30))

print("\nTop 30 Neighborhoods with least bus stops")
print(counts.sort_values(by="point_count", ascending=True).head(30))

counts.sort_values(by="point_count", ascending=False).head(15).plot.bar(x="hood", y="point_count", color = "red")

plt.xlabel("Neighborhoods")
plt.ylabel("Number of Bus Stops")

plt.xticks(rotation=90)
plt.title('Number of Bus Stops per Neighborhood')
plt.show()
```

The top five neighborhoods with the most bus stops are...
1. Central Business District2.  Bloomfiel
3.   Lincoln-Lemington-Belm
4.    Mount Washing
5.     Shadyside 

### Crosswalks per Neighborhood
- How many crosswalks- continental are there in total in this dataset. And in which neighbrohood are the crossswalks in?

Listed the crosswalks-continental and which neighborhood they are i
- After finding out how many crosswalks-continental are there in total in the datasets and which neighborhood it is in.

Found the neighborhood that has the most crosswalks-continental. (Added a column that counts the number of occurrences of each neighborhood)n.

```python
chip = pd.read_csv("Final Project - Crosswalk Data/dataset.tsv", sep="\t")
# Filter for continental crosswalks
continental_crosswalks = chip[chip['type'] == 'Crosswalk - Continental']

# Count the total number of continental crosswalks
total_continental_crosswalks = continental_crosswalks.shape[0]

# List the crosswalks and their neighborhoods
crosswalks_neighborhoods = continental_crosswalks[['type', 'neighborhood']]

# Count the occurrences of each neighborhood
neighborhood_counts = crosswalks_neighborhoods['neighborhood'].value_counts().reset_index()
neighborhood_counts.columns = ['neighborhood', 'count']

# Merge the counts back into the original DataFrame
crosswalks_neighborhoods = crosswalks_neighborhoods.merge(neighborhood_counts, on='neighborhood')

# Sort dataset by decending order
crosswalks_neighborhoods = crosswalks_neighborhoods.sort_values(by="count", ascending=False)

unique_neighborhood_counts = crosswalks_neighborhoods[['neighborhood', 'count']].drop_duplicates().head(15)

# Create the bar graph
unique_neighborhood_counts.plot(kind='bar', x='neighborhood', y='count', figsize=(12, 8), color = "limegreen")
print("Total number of continental crosswalks: ", total_continental_crosswalks)
print(crosswalks_neighborhoods.drop_duplicates())
# Add title and labels
plt.title('Number of Continental Crosswalks per Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Count')
```

The top five neighborhoods with the most crosswalks are...
1. North Oakland
2. Central Business District
3. Shadyside
4. Central Oakland
5. Squirrel Hill North
