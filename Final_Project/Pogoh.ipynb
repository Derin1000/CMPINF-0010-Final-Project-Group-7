```python
def get_total_docks(neighborhood):
    total_docks = pogoh[pogoh['Neighborhoods'] == neighborhood]['Total_Docks'].sum()
    return total_docks

# Standardize all neighborhood names
pogoh['Neighborhoods'] = pogoh['Neighborhoods'].str.strip()

# Then run your total docks code again
Central_Oakland = get_total_docks("Central Oakland")
Downtown = get_total_docks("Downtown")
ShadySide = get_total_docks("ShadySide")  # Correct capitalization if needed
Strip_District = get_total_docks("Strip District")
Bloomfield = get_total_docks("Bloomfield")
South_Side_Flats = get_total_docks("South Side Flats")
South_Shore = get_total_docks("South Shore")
Hazel_Wood = get_total_docks("HazelWood")
Middle_Hill = get_total_docks("Middle Hill")
Homewood_South = get_total_docks("Homewood South")
Larimer = get_total_docks("Larimer")
South_Oakland = get_total_docks("South Oakland")
Central_NorthSide = get_total_docks("Central NorthSide")
Upper_Lawrenceville = get_total_docks("Upper Lawrenceville")
Garfield = get_total_docks("Garfield")
Allegeney_West = get_total_docks("Allegeney West")
CrawFord_Roberts = get_total_docks("CrawFord Roberts")
Homewood_North = get_total_docks("Homewood North")
West_Oakland = get_total_docks("West Oakland")
Lower_Lawrenceville = get_total_docks("Lower Lawrenceville")
East_Liberty = get_total_docks("East Liberty")
North_Shore = get_total_docks("North Shore")
Spring_Hill = get_total_docks("Spring Hill")
Chateau = get_total_docks("Chateau")
Point_breeze_North = get_total_docks("Point breeze North")


# Print the results
print(f"Total docks in Central Oakland: {Central_Oakland}")
print(f"Total docks in Downtown: {Downtown}")
print(f"Total docks in ShadySide: {ShadySide}")
print(f"Total docks in Strip District: {Strip_District}")
print(f"Total docks in Bloomfield: {Bloomfield}")
print(f"Total docks in South Side Flats: {South_Side_Flats}")
print(f"Total docks in South Shore: {South_Shore}")
print(f"Total docks in HazelWood: {Hazel_Wood}")
print(f"Total docks in Middle Hill: {Middle_Hill}")
print(f"Total docks in Homewood South: {Homewood_South}")
print(f"Total docks in Larimer: {Larimer}")
print(f"Total docks in South Oakland: {South_Oakland}")
print(f"Total docks in Central_NorthSide: {Central_NorthSide}")
print(f"Total docks in Upper Lawrenceville: {Upper_Lawrenceville}")
print(f"Total docks in Garfield: {Garfield}")
print(f"Total docks in Allegeney West: {Allegeney_West}")
print(f"Total docks in CrawFord Roberts: {CrawFord_Roberts}")
print(f"Total docks in Homewood North: {Homewood_North}")
print(f"Total docks in West Oakland: {West_Oakland}")
print(f"Total docks in Lower Lawrenceville: {Lower_Lawrenceville}")
print(f"Total docks in East Liberty: {East_Liberty}")
print(f"Total docks in North Shore: {North_Shore}")
print(f"Total docks in Spring Hill: {Spring_Hill}")
print(f"Total docks in Chateau: {Chateau}")
print(f"Total docks in Point breeze North: {Point_breeze_North}")
```

```python
# Group by neighborhoods and calculate total docks for each
neighborhood_totals = pogoh.groupby('Neighborhoods')['Total_Docks'].sum()

# Find the neighborhood with the maximum total docks
max_neighborhood = neighborhood_totals.idxmax()
max_total_docks = neighborhood_totals.max()
print(f"The neighborhood with the most total docks is {max_neighborhood} with {max_total_docks} docks.")
```

```python
import matplotlib.pyplot as plt

neighborhood_docks = pogoh.groupby('Neighborhoods')['Total_Docks'].sum().reset_index()

courses = neighborhood_docks['Neighborhoods']
values = neighborhood_docks['Total_Docks']

plt.bar(courses, values, color='darkblue', width=0.4)

plt.xlabel("Neighborhoods")
plt.ylabel("Total No of Docks")

plt.xticks(rotation=90)

plt.show()
```

```python
import matplotlib.pyplot as plt

neighborhood_docks = pogoh.groupby('Neighborhoods')['Total_Docks'].sum().nlargest(5).reset_index()

courses = neighborhood_docks['Neighborhoods']
values = neighborhood_docks['Total_Docks']

plt.bar(courses, values, color='purple', width=0.4)

plt.xlabel("Neighborhoods")
plt.ylabel("Total No of Docks")

plt.xticks(rotation=90)

plt.show()
```
